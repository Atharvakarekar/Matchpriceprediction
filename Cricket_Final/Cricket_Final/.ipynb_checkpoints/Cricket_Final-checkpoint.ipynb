```python
import pandas as pd
import numpy as np
```

```python
icc = pd.read_csv("IccDiffData.csv")
icc
```

```python

icc.head()
```

```python
icc["Team1"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Team1 = LabelEncoder()
icc['Team1'] = le_Team1.fit_transform(icc['Team1'])
icc['Team1'].unique()

```

```python
icc["Team2"].unique()
```

```python
le_Team2 = LabelEncoder()
icc['Team2'] = le_Team2.fit_transform(icc['Team2'])
icc['Team2'].unique()
```

```python
X= icc[['Team1','Team2' ,'Rating1','Rating2','RatingDiff']]
y=icc[['Stand1','Stand2']]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.1)
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)
```

```python
y_pred= linear_reg.predict(X_test)
y_pred
```

```python
linear_reg.score(X_test, y_test)
```

```python
import pickle
```

```python
data = {"model":linear_reg, "le_Team1": le_Team1, "le_Team2" : le_Team2 }
with open('cricket_pred.pkl','wb') as file:
    pickle.dump(data,file)
```

```python
with open('cricket_pred.pkl','rb') as file:
    data= pickle.load(file)

model = data["model"]
le_Team1= data["le_Team1"]
le_Team2= data["le_Team2"]
    
```
